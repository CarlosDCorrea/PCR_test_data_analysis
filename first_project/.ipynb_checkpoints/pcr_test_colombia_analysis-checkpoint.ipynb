## 1.1 Data Cleaning

### after we visualice our data, we notice that there are some missing values in it, this missing vallues are in all the row without losing information and therefore we can delete the entire row. 

### Note: there are different ways in wich we can deal with missing data, in this case we will drop the rows but we also can deal with them by imputing the missing values or predicting them.

```python
cleaned_data = df.dropna()
```

```python
cleaned_data
```

### We really dont need the format that the column "Fecha" has currently, hence i gonna change it to the formar "yyyy-MM-dd"

```python
def format_date(date):
    return date[0:10]
```

```python
cleaned_data['Fecha'] = cleaned_data['Fecha'].apply(format_date)
```

```python
cleaned_data.describe()
```

### In our data set we have de column named "Positividad acumulada" which is the porcentage of the positive samples with all the samples, what we gonna do is to create two new columns with the porcentage of negativity and indeterminity samples.

```python
negatividad_acumulada = round((cleaned_data['Negativas acumuladas'] / cleaned_data['Acumuladas'].values)*100, 1)
```

```python
indeterminidad_acumulada = round((cleaned_data['Indeterminadas'] / cleaned_data['Acumuladas'].values)*100, 1)
```

```python
cleaned_data.insert(41, "Negatividad acumulada", negatividad_acumulada.values, True)
```

```python
cleaned_data.insert(42, "Indeterminidad acumulada", indeterminidad_acumulada.values, True)
```

## 1.2 Data Visualization

```python
plt.figure(figsize=(10,5))
plt.title("acumulated PCR test")
cleaned_data['Acumuladas'].plot().grid()
```

```python
def visualice_by_closes_department(df, closes_departments):
    plt.figure(figsize=(17, 5))
    plt.title('PCR test by department')
    plt.ylabel("made test")
    for col in closes_departments:
        df[col].plot(label = col).grid(True)

    plt.legend()
```

```python
closes_department1 = ['Narino', 'Cauca', 'Putumayo', 'Amazonas', 'Caqueta']
visualice_by_closes_department(cleaned_data, closes_department1)
```

```python
closes_department2 = ['Meta', 'Casanare', 'Boyaca', 'Arauca', 'Cundinamarca', 'Huila']
visualice_by_closes_department(cleaned_data, closes_department2)
```

```python
closes_department3 = ['Choco', 'Antioquia', 'Caldas', 'Risaralda', 'Cordoba']
visualice_by_closes_department(cleaned_data, closes_department3)
```

```python
closes_department4 = ['Cesar', 'Magdalena', 'Atlantico', 'Sucre', 'Bogota']
visualice_by_closes_department(cleaned_data, closes_department4)
```

### The idea was to separate the departments taking into account if they are close to each other, in that way we can determinate if the number of PCR tests made depends in the closely of the departments, for instance, BogotÃ¡ is a city with a big population, thus we can infer that the infected people is more than in other departments, but also it might mean that the departments closer to BogotÃ¡ may have increase in virus infections and therefore the PCR tests in these departments increase too, but looking at the graphs it is not like that, at the end the amount of PCR tests depends in the population.

```python
plt.figure(figsize=(15, 7))
cleaned_data['Positivas acumuladas'].plot(color = 'g', label = 'Positivas acumuladas')
cleaned_data['Negativas acumuladas'].plot(color = 'r', label = 'Negativas acumuladas')
plt.legend()
plt.figure(figsize=(15, 7))
cleaned_data['Indeterminadas'].plot(color = 'b')
```

### in the above plots we see the positive, negative and indeterminate PCR test, as expected, the negative PCRT tests are more than the positive ones which is good news but it could be better.

```python
rcParams['figure.figsize'] = 18, 10
decomposition = sm.tsa.seasonal_decompose(cleaned_data['Positivas acumuladas'], model='additive', freq=12)
fig = decomposition.plot()
plt.show()
```

### The image above show us the trend, observation and the seasonality of the positive PCR test, we can see that it is an upward trend which is bad, also the seasonality which is the frecuency of the trend in a period of time is stable, it means that the uptrend wont break up in the next months.

porcentage of positives, negatives and indeterminity.

```python
means = round(cleaned_data.iloc[:, -4:-1].mean(), 2)
labels = ['Negatividad', 'Positividad', 'Indeterminidad']
colors = ["r", "g", "b"]
explode = [0, 0.1, 0]
pct = '%.1f%%'
```

```python
plt.pie(means, labels=labels, colors=colors, autopct=pct, explode=explode)
plt.title("PCR test porcentage")
plt.show()
```

### Finally we can see that the positivity of the PCR tests is very low compare with the negativity, the idea is that the positivity of the PCR tests must get lower and the government and the society are responsible of it.
